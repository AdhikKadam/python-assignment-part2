```python
# --- Task 1.1: Print the full menu grouped by category ---

# Extract unique categories in the order they appear
categories = []
for details in menu.values():
    if details["category"] not in categories:
        categories.append(details["category"])

for category in categories:
    print(f"===== {category} =====")
    # Loop through the menu and filter by the current category
    for item_name, details in menu.items():
        if details["category"] == category:
            status = "[Available]" if details["available"] else "[Unavailable]"
            # Format: <17 ensures 17 character spacing for names, <7 ensures 7 spacing for price including 2 decimal places
            print(f"{item_name:<17} ₹{details['price']:<7.2f} {status}")
    print() # Add a blank line between categories
```

```python
# --- Task 1.2: Dictionary computations and statistics ---
# Find available items count
total_items = len(menu)
available_items = sum(1 for details in menu.values() if details["available"])

# Find the most expensive item using max() with a custom key
expensive_name, expensive_details = max(menu.items(), key=lambda x: x[1]["price"])
expensive_price = expensive_details["price"]

# Find all items under ₹150
items_under_150 = [(name, details["price"]) for name, details in menu.items() if details["price"] < 150]

print("--- Menu Statistics ---")
print(f"Total number of items: {total_items}")
print(f"Total available items: {available_items}")
print(f"Most expensive item:   {expensive_name} (₹{expensive_price:.2f})")
print("Items under ₹150:")
for name, price in items_under_150:
    print(f"  - {name} (₹{price:.2f})")
```

# Task 2 - Cart Operations 

```python
# Initial empty cart
cart = []

# --- Logic Functions ---

def print_cart(action_message):
    """Helper function to print the current state of the cart."""
    print(f"\n--- {action_message} ---")
    if not cart:
        print("Cart is empty.")
    else:
        for item in cart:
            print(item)

def add_item(item_name, quantity):
    # 1. Check if the item exists in the menu
    if item_name not in menu:
        print(f"\n❌ Cannot add '{item_name}': Item does not exist on the menu.")
        return
    
    # 2. Check if the item is available
    if not menu[item_name]["available"]:
        print(f"\n❌ Cannot add '{item_name}': Item is currently unavailable.")
        return
        
    # 3. Check if item is already in the cart to update quantity
    for cart_item in cart:
        if cart_item["item"] == item_name:
            cart_item["quantity"] += quantity
            print_cart(f"Updated '{item_name}' quantity by +{quantity}")
            return
            
    # 4. If all checks pass and it's not in the cart, add as a new entry
    cart.append({
        "item": item_name,
        "quantity": quantity,
        "price": menu[item_name]["price"]
    })
    print_cart(f"Added {quantity}x '{item_name}' to cart")

def remove_item(item_name):
    # Iterate through a copy of the list or use enumeration to remove safely
    for i in range(len(cart)):
        if cart[i]["item"] == item_name:
            removed = cart.pop(i)
            print_cart(f"Removed '{item_name}' from cart")
            return
            
    # If the loop finishes without returning, the item wasn't found
    print(f"\n❌ Cannot remove '{item_name}': Not found in the cart.")

def update_quantity(item_name, new_quantity):
    if new_quantity <= 0:
        print(f"\n⚠️ Quantity must be greater than 0. Use remove_item instead.")
        return
        
    for cart_item in cart:
        if cart_item["item"] == item_name:
            cart_item["quantity"] = new_quantity
            print_cart(f"Set '{item_name}' quantity to {new_quantity}")
            return
            
    print(f"\n❌ Cannot update '{item_name}': Not found in the cart.")


# --- Simulation Sequence ---

# 1. Add "Paneer Tikka" × 2
add_item("Paneer Tikka", 2)

# 2. Add "Gulab Jamun" × 1
add_item("Gulab Jamun", 1)

# 3. Add "Paneer Tikka" × 1 -> quantity should update to 3
add_item("Paneer Tikka", 1)

# 4. Try to add "Mystery Burger" -> does not exist
add_item("Mystery Burger", 1)

# 5. Try to add "Chicken Wings" -> exists but unavailable
add_item("Chicken Wings", 2)

# 6. Remove "Gulab Jamun"
remove_item("Gulab Jamun")


# --- Final Order Summary ---

print("\n\n========== Order Summary ==========")
subtotal = 0.0

for item in cart:
    item_total = item["quantity"] * item["price"]
    subtotal += item_total
    # Formatting: <18 for left-aligned name, >2 for right-aligned qty, >7 for right-aligned price
    print(f"{item['item']:<18} x{item['quantity']:<4} ₹{item_total:>7.2f}")

print("------------------------------------")

# Calculate taxes and final total
gst = subtotal * 0.05
total_payable = subtotal + gst

print(f"{'Subtotal:':<25} ₹{subtotal:>7.2f}")
print(f"{'GST (5%):':<25} ₹{gst:>7.2f}")
print(f"{'Total Payable:':<25} ₹{total_payable:>7.2f}")
print("====================================")
```

# Task 3 - Inventory Tracker with Deep Copy 

```python
import copy

# Initial Data
inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}

# The final cart state from the previous task
cart = [
    {"item": "Paneer Tikka", "quantity": 3, "price": 180.0}
]

# --- 1. Deep Copy and Demonstration ---
print("--- Deep Copy Demonstration ---")

# Deep copy is required here because 'inventory' contains nested dictionaries.
# A standard .copy() would only copy the top-level keys, leaving the inner dicts linked.
inventory_backup = copy.deepcopy(inventory)

# Modify the original to prove they are separate
inventory["Paneer Tikka"]["stock"] = 999

print(f"Modified Original  (Paneer Tikka stock): {inventory['Paneer Tikka']['stock']}")
print(f"Unaffected Backup  (Paneer Tikka stock): {inventory_backup['Paneer Tikka']['stock']}")

# Restore the original from the backup to continue the simulation
inventory = copy.deepcopy(inventory_backup)
print("✓ Inventory restored to original state.\n")


# --- 2. Simulate Order Fulfillment ---
print("--- Order Fulfillment ---")

for cart_item in cart:
    item_name = cart_item["item"]
    qty_needed = cart_item["quantity"]
    
    if item_name in inventory:
        current_stock = inventory[item_name]["stock"]
        
        if current_stock >= qty_needed:
            # Sufficient stock, deduct normally
            inventory[item_name]["stock"] -= qty_needed
            print(f"✓ Deducted {qty_needed}x '{item_name}' from inventory.")
        else:
            # Insufficient stock, deduct what is available
            print(f"⚠️ Warning: Insufficient stock for '{item_name}'. "
                  f"Needed {qty_needed}, but only {current_stock} available.")
            inventory[item_name]["stock"] = 0
            print(f"✓ Deducted {current_stock}x '{item_name}' from inventory (stock now empty).")


# --- 3. Reorder Alerts ---
print("\n--- Reorder Alerts ---")
alerts_triggered = False

for item, details in inventory.items():
    if details["stock"] <= details["reorder_level"]:
        print(f"⚠ Reorder Alert: {item} — Only {details['stock']} unit(s) left (reorder level: {details['reorder_level']})")
        alerts_triggered = True

if not alerts_triggered:
    print("✓ All items are currently above their reorder levels.")


# --- 4. Final Proof of Deep Copy Protection ---
print("\n--- Final State Check (Original vs Backup) ---")
print(f"{'Item Name':<15} | {'Current Stock':<13} | {'Backup Stock'}")
print("-" * 47)

for item in inventory:
    current = inventory[item]["stock"]
    backup = inventory_backup[item]["stock"]
    # Mark rows that have changed with a '*' for easy reading
    changed = "*" if current != backup else ""
    print(f"{item:<15} | {current:<13} | {backup} {changed}")
```

# Task 4 — Daily Sales Log Analysis

```python
# Helper function to print revenue stats so we can reuse it
def print_revenue_stats(log_data):
    best_day = ""
    max_revenue = 0.0
    
    for date, orders in log_data.items():
        # Calculate daily revenue by summing the 'total' of each order
        daily_revenue = sum(order["total"] for order in orders)
        print(f"  {date}: ₹{daily_revenue:.2f}")
        
        # Check if this is the best day so far
        if daily_revenue > max_revenue:
            max_revenue = daily_revenue
            best_day = date
            
    print(f"\n  🏆 Best-selling day: {best_day} (₹{max_revenue:.2f})\n")


# --- Task 4.1 & 4.2: Initial Revenue & Best-Selling Day ---
print("===== Initial Revenue Per Day =====")
print_revenue_stats(sales_log)


# --- Task 4.3: Most Ordered Item ---
item_frequency = {}

for orders in sales_log.values():
    for order in orders:
        for item in order["items"]:
            # Increment the count for the item, defaulting to 0 if it doesn't exist yet
            item_frequency[item] = item_frequency.get(item, 0) + 1

# Find the item with the highest count
most_ordered_item = max(item_frequency, key=item_frequency.get)
highest_count = item_frequency[most_ordered_item]

print("===== Item Popularity =====")
print(f"Most ordered item: {most_ordered_item} ({highest_count} orders)\n")


# --- Task 4.4: Add New Day and Reprint Stats ---
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

print("===== Updated Revenue Per Day (After 2025-01-05) =====")
print_revenue_stats(sales_log)


# --- Task 4.5: Numbered List of All Orders ---
print("===== All Orders History =====")

# First, flatten the dictionary into a list of tuples containing (date, order_dict)
all_orders_flat = []
for date, orders in sales_log.items():
    for order in orders:
        all_orders_flat.append((date, order))

# Now use enumerate to cleanly number them starting from 1
for index, (date, order) in enumerate(all_orders_flat, start=1):
    # Join the items list into a clean comma-separated string
    items_str = ", ".join(order["items"])
    print(f"{index}.  [{date}] Order #{order['order_id']:<2} — ₹{order['total']:<6.2f} — Items: {items_str}")
```
